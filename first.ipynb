```python
f2 = iio.imread('draft.jpg')
plt.imshow(f2, cmap=plt.cm.gray)
```

```python
print(f2.shape),
f2.dtype
```

```python
victim = np.copy(f2)
new = np.empty([532,700],dtype = np.uint8)
for i in range(len(victim)):
    for j in range(len(victim[i])):
        if victim[i][j] > 128:
            new[i][j]=255
plt.imshow(new, cmap=plt.cm.gray)
```

```python
f3 = iio.imread('pika.jpg')
plt.imshow(f3)
plt.show()
```

```python
f3.dtype
```

```python
#Thresholds
# [RLT, RUT, GLT,GHT,BLT,BHT]
cthresh = [200,55,0,55,180,255]
```

```python
new2 = np.empty([412,474], dtype = np.uint8)
for i in range(len(f3)):
    for j in range(len(f3[i])):
        rgbpix = f3[i][j]
        if rgbpix[0]>cthresh[0] and rgbpix[0]<cthresh[1]:
            if rgbpix[1]>cthresh[2] and rgbpix[1]<cthresh[3]:
                if rgbpix[2]>cthresh[4] and rgbpix[2]<cthresh[5]:
                    new2[i][j]=255
```

```python
plt.imshow(new2, cmap=plt.cm.gray)
```
