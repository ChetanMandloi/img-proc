```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(18,6))

for c, ax in zip(range(3), axs):
    tmp_img = np.zeros(img.shape, dtype="uint8")
    tmp_img[:,:,c] = img[:,:,c]
    ax.imshow(tmp_img)
    ax.set_axis_off()
```

```python
img2 = np.copy(img)
def brighten(image, factor):
    #print(image,'b')
    for i in range(len(image)):
        for j in range(len(image[i])):
            for k in range(len(image[i][j])):
                image[i][j][k] = min(255,image[i][j][k]*factor)
    #print(image)
    return image
img2 = brighten(img2,1.3)
```

```python
plt.imshow(img)
plt.show()
plt.imshow(img2)
plt.show()
```
